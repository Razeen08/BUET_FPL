# Insert League Name

```python
league_name = 'BUET FPL Classic'
```

```python
import os
import warnings
warnings.filterwarnings("ignore")

from urllib.request import urlopen
import json
import pandas as pd
import numpy as np
from IPython.display import display, HTML

!pip install xlsxwriter
```

```python
url_part1 = 'https://fantasy.premierleague.com/api/leagues-classic/'
url_part3 = '/standings/?page_standings='
df = pd.DataFrame()
for page in np.arange(1,200000):
    url = url_part1 + str(league_id) + url_part3 + str(page)
    response = urlopen(url)
    data_json = json.loads(response.read())
    df_temp = pd.DataFrame(data_json['standings']['results'])
    df = pd.concat([df,df_temp])
    if data_json['standings']['has_next'] == False:
        break

file = league_name+'.xlsx'
writer = pd.ExcelWriter(file, engine='xlsxwriter')
sheet_name = 'League Data GW' + str(GW)
df.to_excel(writer, sheet_name=sheet_name, index=False)
writer.save()
df.set_index(np.arange(len(df)))

df = df.sort_values(by=['event_total', 'rank', 'entry'], ascending=[False, True, True])

motw_url = 'https://fantasy.premierleague.com/entry/' + str(df['entry'].iloc[0]) + '/event/' + str(GW)
link_text = league_name + ' MOTW GW' + str(GW)
font_size = "32px"

link_html = f'<a href="{motw_url}" target="_blank" style="font-size: {font_size};">{link_text}</a>'
display(HTML(link_html))
```
